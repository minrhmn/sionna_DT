```python
import matplotlib.pyplot as plt

# Define the SNR and PER values
snr = list(range(20, 29))  # SNR values from 20 to 28
per = [0.998, 0.984, 0.886, 0.65, 0.326, 0.141, 0.062, 0.023, 0.001]

# Create a figure
plt.figure(figsize=(8, 6))

# Plot the data with a logarithmic scale on the y-axis
plt.semilogy(snr, per, marker='o', linestyle='-', color='b', label='PER vs SNR')

# Label the axes and add a title
plt.xlabel('SNR (dB)')
plt.ylabel('PER')
plt.title('Packet Error Rate (PER) vs Signal-to-Noise Ratio (SNR)')

# Enable grid lines for better readability
plt.grid(True, which="both", ls="--")

# Add a legend
plt.legend()

# Display the plot
plt.show()
```
